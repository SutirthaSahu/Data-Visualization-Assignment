```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris_data  = load_iris()
```

```python
df_iris = pd.DataFrame(data=np.c_[iris_data['data'],iris_data['target']],columns=iris_data['feature_names']+['target'])
df_iris
```

```python
X = df_iris.iloc[:,:-1].values
y = df_iris.iloc[:,1].values
```

```python
#Label Encoding target values

from sklearn.preprocessing import OneHotEncoder,LabelEncoder

lblEncoder = LabelEncoder()
y = lblEncoder.fit_transform(y)
```

```python
#Feature Scaling

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
#Reducing dimension to 3D

from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X = pca.fit_transform(X)
```

```python
#Visualizing 3D chart

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(1, figsize=(4,3))
plt.clf()
ax = Axes3D(fig, rect=[1,2,2,3], elev=60, azim=150)

ax.scatter(X[:,0], X[:,1], X[:,2], c=y , edgecolor='k')
plt.show()
```
